```python
import cv2
```

An image inpainting technique based on the fast marching method (Telea, 2004)

```python
# initialize the inpainting algorithm to be the Telea et al. method
flags = cv2.INPAINT_TELEA
```

```python
dir = '../data/'
image = cv2.imread(dir+'impainting_image.png',1)
mask = cv2.bitwise_not(cv2.imread(dir+'impainting_mask.png'))
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
# perform inpainting using OpenCV
output = cv2.inpaint(image, mask, inpaintRadius=3, flags=flags)
```

```python
# show the original input image, mask, and output image after
# applying inpainting
import matplotlib.pyplot as plt
# load image using cv2....and do processing.
print('Remakr: OpenCV uses BGR as its default color order,\n '
      'when you wanna use matplotlib to visualize it, \n'
      'You need add additional cv2.COLOR_BGR2RGB.')

plt.figure(figsize = (20,10))
plt.subplot(311)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.subplot(312)
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))
plt.subplot(313)
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.show()
```

Navier-stokes, Fluid dynamics, and image and video inpainting (Bertalmo et al., 2001)

```python
flags = cv2.INPAINT_NS
dir = '../data/'
image = cv2.imread(dir+'impainting_image.png',1)
mask = cv2.bitwise_not(cv2.imread(dir+'impainting_mask.png'))
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
# perform inpainting using OpenCV
output = cv2.inpaint(image, mask, inpaintRadius=3, flags=flags)
# show the original input image, mask, and output image after
# applying inpainting
import matplotlib.pyplot as plt
# load image using cv2....and do processing.
print('Remakr: OpenCV uses BGR as its default color order,\n '
      'when you wanna use matplotlib to visualize it, \n'
      'You need add additional cv2.COLOR_BGR2RGB.')

plt.figure(figsize = (20,10))
plt.subplot(311)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.subplot(312)
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))
plt.subplot(313)
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.show()
```

```python
fig = plt.figure(figsize=(19.20,10.80),frameon=False)
ax = plt.Axes(fig, [0., 0., 1., 1.])
ax.set_axis_off()
fig.add_axes(ax)
ax.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
fig.savefig('../data/impainting_out.png', dpi=600)
```
